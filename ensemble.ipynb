#Ensemble

We take our two best models, XGBoost and Catboost, and setup a basic ensemble by blending their predictions. Our final milk yield is a weighted average (60% Catboost, 40% XGBoost).

```python
import pandas as pd

cat = pd.read_csv("submission_catboost.csv")
xgb = pd.read_csv("submission_xgboost.csv")

merged = pd.merge(cat, xgb, on="Cattle_ID", suffixes=("_cat", "_xgb"))
merged["Milk_Yield_L"] = (
    0.6 * merged["Milk_Yield_L_cat"] +
    0.4 * merged["Milk_Yield_L_xgb"]
)
final = merged[["Cattle_ID", "Milk_Yield_L"]]
final.to_csv("submission_ensemble.csv", index=False)

print(final.head())

print("\nSaved final ensemble submission to submission_ensemble.csv")


```
